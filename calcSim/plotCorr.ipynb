```python
import pandas as pd
import matplotlib.pyplot as plt
csv = './sum.csv'
df = pd.read_csv(csv)
df = df.drop(['name','index'], axis=1)
```

```python
df.corr(method='spearman').style.format("{:.2}").background_gradient(cmap=plt.get_cmap('coolwarm'), axis=1)
```

```python
df['top1-enwiki'].corr(df['val_loss'])
```

```python
df.corr(method='spearman').style.format("{:.2}").background_gradient(cmap=plt.get_cmap('coolwarm'), axis=1)
```

```python
df.corr(method='kendall').style.format("{:.2}").background_gradient(cmap=plt.get_cmap('coolwarm'), axis=1)
```

```python
df.corr(method='pearson').style.format("{:.2}").background_gradient(cmap=plt.get_cmap('coolwarm'), axis=1)
```

```python
# plot correlated values
plt.style.use('ggplot')
plt.rcParams['figure.figsize'] = [20, 5]

fig, ax = plt.subplots(nrows=1, ncols=4)

ax=ax.flatten()

cols = ['top1-simpwiki', 'top20-simpwiki', 'top50-simpwiki','top100-simpwiki']
colors=['red', 'blue', 'purple',
                 'green', '#88BB44', '#FFBBBB']
j=0

for i in ax:
    if j==0:
        i.set_ylabel('val loss')
    i.scatter(df[cols[j]], df['val_loss'],  alpha=0.5, color=colors[j])
    i.set_xlabel(cols[j])
    i.set_title('Pearson: %s'%df.corr().loc[cols[j]]['val_loss'].round(2)+' Spearman: %s'%df.corr(method='spearman').loc[cols[j]]['val_loss'].round(2))
    j+=1
    i.grid()

plt.show()
```
